```python
def remove_link(row):
    for r in [x for x in re.findall(r"\((.*?)\)", row) if re.search(r"\d\d\.\d\d\.\d\d\d\d", x)]:
        row = row.replace(f"({r})", "")
    return row
```

```python
df.text = df.text.apply(lambda x: remove_link(x))
```
